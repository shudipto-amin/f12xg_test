## `get_cbs_from_values`

For `HF=True`, should follow:

$$
A + B e^{-X}
$$

For 'HF=False`, should follow:

$$
A + B X^{-exp}
$$

```python
X = np.array([2, 3, 4, 5])
```

### Test `HF=True`

```python
reload(ffa)
A = -3
B = 1.5
energies = A + B * np.exp(-X)
print(energies)
ffa.get_cbs_from_values(X, energies, HF=True)
```

### Test `HF=False`

```python
reload(ffa)
exp = 3
A = -4.0
B = 1.5
energies = A + B * X**(-exp)
ffa.get_cbs_from_values(X, energies, exp=exp) == A
```
